```python
import numpy as np
import numpy.random as npr
import pandas as pd
import time 
```

```python
class Plat : 
    def __init__(self ,ide , nom , pmin , pmax , stock):
        self.id = ide 
        self.nom = nom 
        self.pmin = pmin
        self.pmax = pmax 
        self.stock = stock 
        pas = (pmax - pmin) / (k - 1)
        self.liste_prix = [(pmin + i * pas) for i in range(k)]
    
    def afficher(self):
        print("id : " , self.id , "\nnom: " , self.nom , "\n(pmin , pmax) : (", self.pmin , "  , " , self.pmax , ")\nstock : " , self.stock )
```

```python
#Creation p plats 
Np = 10 
k = 10
list_plats = []
for i in range(Np):
    nom = "plat_" + str(i+1)
    ide = i+1 
    pmin = npr.randint(5 , 11)
    pmax = npr.randint(20 , 31)  
    stock = npr.randint(80 , 121)
    list_plats.append(Plat(ide , nom , pmin , pmax , stock))
for pl in list_plats : 
    pl.afficher()
```

```python
from random import * 
DEB = time.time()
#Creation DataFrame 
Df = pd.DataFrame(columns = [ "Client" , "Classa" ,"Classf" ,"Mois" ,"Jour" , "Periode" , "Plat" , "Note" , "Prix" , "Achat" ])

#Constantes de simulations 
#Nbre d'utilisateurs au départ 
N_utilisateurs = 10000
l_clients = range(N_utilisateurs)#Liste des clients 
#Nbre de mois , de jours et de periodes 
Nm = 1 
Nj = 30
nbre_etats = 24 

#Differentes classes des clients et leur proportion moyen 
Class_f  = {"pauvre" : 1/4 , "moyen" : 1/2 , "riche" : 1/4}
Class_a  = {"aime_pas" : 1/4 , "indifferent" : 1/2 , "aime" : 1/4}

#Differentes classes des jours et periodes 
Class_j = {"debut":list(range(3 , 10)) , "milieu" : list(range(10 , 25)) , "fin" :[1 , 2] + list(range(25 , Nj+1))}
Class_p = {"debut":list(range(6)) , "milieu" : list(range(6 , 18)) , "fin" :list(range(18 , nbre_etats))}

#Differentes proportion de chaque classe (pauvre , moyen , riche )
prop_seg_f = npr.multinomial(N_utilisateurs , list(Class_f.values())) 
liste_prop_seg_f = []
temp = l_clients
for i in range(len(prop_seg_f)):
    liste_prop_seg_f.append(sample(temp ,prop_seg_f[i] ))
    temp = [g for g in temp if g not in liste_prop_seg_f[i]]
print(len(liste_prop_seg_f[0]) , " " , len(liste_prop_seg_f[1]) , " " , len(liste_prop_seg_f[2]) , " " )
#Liste des noms des plats 
lplnom = []
for plat in list_plats :
    lplnom.append(plat.nom)
    
Class_j_vals = []
Class = [Class_f , Class_a]

for mois in range( 1 , Nm+1):  #Boucle dans un mois 
    
    
    
    print("debut du mois : " , mois)
    lamda = npr.uniform(20, 30) #Nbre moyen de client par période dans ce mois 
    prix_pauvre_m = npr.uniform(5, 15) #propension moyen à payer pour le pauvvre
    prix_moyen_m = npr.uniform(1.5, 2) * prix_pauvre_m #propension moyen à payer pour le moyen
    prix_riche_m = npr.uniform(2, 3) * prix_moyen_m #propension moyen à payer pour le riche
    Class_prix_m = [prix_pauvre_m , prix_moyen_m ,  prix_riche_m]
    
    
    
    
    for jour in range(1 , Nj+1): #Boucle dans un jour 
        
        
        deb = time.time() 
        print("Debut du jour : " , jour)
        lamda2 = npr.poisson(lamda) #Nbre moyen de client par période dans la journée en cours 
        Class_prix_j =[]        
        for cpt in range(len(Class_prix_m)):            
            Class_prix_j.append(npr.poisson(Class_prix_m[cpt])) #Liste contenant les propensions à payer des differentes classes durant cette journée 
            
        #if jour in Class_j["debut"]:
        if (jour>=3 and jour<10): #Si debut de mois , propension un peu élevée 
            Class_prix_j = [npr.uniform(1 , 1.5)*temp for temp in Class_prix_j]
            
        #if jour in Class_j["fin"]:
        if (jour>=25 or jour <3): #Si fin du mois , propension encore plus élevée 
            Class_prix_j = [npr.uniform(1 , 2)*temp for temp in Class_prix_j]
                        
        liste_clients_r =  l_clients #liste des clients suceptibles d'acheter .... possibilité que quelqu un ayant acheté le fasse encore 
        
        for periode in range(nbre_etats): #Boucle dans une periode 
            
            Class_prix_p =[] #liste des prix dans cette periode 
            for cpt in range(len(Class_prix_j)):            
                Class_prix_p.append(npr.poisson(Class_prix_j[cpt]))
                
            if periode in Class_p["milieu"]: #Si milieu du mois , propension élevée 
                Class_prix_p = [npr.uniform(1 , 2)*temp for temp in Class_prix_p]
                
            if periode in Class_p["fin"]: #Si fin du mois , propension encore élevée mais moins 
                Class_prix_p = [npr.uniform(1 , 1.5)*temp for temp in Class_prix_p]
                
            #Dictionnaire des prix pour chaque plat 
            Dic_prix = {}
            for plat in list_plats : 
                Dic_prix[plat] = npr.choice(plat.liste_prix)
                
                                
                
            N = npr.poisson(lamda2) #Nbre de client à la période en cours 
            liste_clients_p = sample(liste_clients_r ,N ) #liste des clients visitant la plateforme à la période en cours 
            liste_clients_r = [i for i in liste_clients_r if not i in liste_clients_p] #Actualisation de la liste possible     NB....
            
            for client in liste_clients_p: #Boucle sur les clients .... encours 
                #DataFrame pour chaque client 
                dp = pd.DataFrame(columns = [ "Mois" , "Jour","Periode","Client" ,"Classf", "Classa" , "Plat","Note" ,"Prix" ,"Propension" , "Achat"]) 
                #classe financiere 
                ind_f = 0 
                for ind in range (len(Class_f)): #Indice financier pauvre:0 , moyen:1 ....
                    
                    if client in liste_prop_seg_f[ind]:
                        ind_f = ind 
                        
                Prix = list(Dic_prix.values()) #Liste des prix Choisis 
                temp_a = []
                Dic_proppay = [] #Liste Propension à  payer pour chaque plat
                propmax = 0 
                Note = [] #Liste des notes 
                for ind in range(Np):#Boucle sur chaque plat 
                     #Propension à payer 
                    propension = Class_prix_p[ind_f]
                   
                    cpt_a = choices(range(len(Class_a)) , list(Class_a.values())) #choix , pour chaque client de la Classe aime ,....
                    cpt_a = cpt_a[0]
                    if cpt_a == 0 : #Aime pas 
                        note = npr.randint(1 , 4)
                    if cpt_a ==1 :  #Indifferent 
                        note = npr.randint(1 , 6)
                    if cpt_a ==2 : #Aime
                        note = npr.randint(3 , 6) 
                    Note.append(note)
                    propension = propension*(note/3)
                    Dic_proppay.append(propension)
                    
                    temp_a.append(list(Class_a.keys())[cpt_a])
                    if (propmax < (propension / Dic_prix[plat]) ) : #si on achete , alors on achete le produit qui a le plus grand rapport propension a payer / prix 
                        propmax = propension / Dic_prix[plat]                         
                        platmax = plat
                        indmax = ind 
                Achat = [0 for i in range(Np)]
                #On achete ssi la propension à payer est plus grande que le prix 
                if (propmax >= 1): 
                    Achat[indmax] = 1
                
                temp = {"Mois" : mois , "Client" : client , "Classf" : list(Class_f.keys())[ind_f], "Classa" : temp_a ,"Propension": Dic_proppay ,"Jour":jour , "Periode" : periode ,"Plat" : lplnom ,"Note" : Note , "Prix" : Prix ,"Achat" : Achat }
                dp =pd.DataFrame(temp)                   
                Df = pd.concat([Df ,dp])
            
        print("durée : " , ,time.time() - deb)
print(time.time() - DEB)            
            
            
```
